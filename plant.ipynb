```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import os
```

```python
DATASET_DIR = "/content/plantvillage/plantvillagedataset/color"
```

```python
IMG_SIZE = (128, 128)
BATCH_SIZE = 32
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=25,
    zoom_range=0.2,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    horizontal_flip=True,
    validation_split=0.2
)

test_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)
```

```python
train_data = train_datagen.flow_from_directory(
    DATASET_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    subset="training"
)
```

```python
val_data = test_datagen.flow_from_directory(
    DATASET_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    subset="validation"
)
```

```python
num_classes = train_data.num_classes
print("Number of classes:", num_classes)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(128,128,3)),
    tf.keras.layers.MaxPool2D(),

    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPool2D(),

    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPool2D(),

    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(num_classes, activation='softmax')
])
```

```python
model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()
```

```python
history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=10
)
```

```python
plt.figure(figsize=(12,4))

# Accuracy
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title("Training & Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend(["Train", "Validation"])
```

```python
plt.subplot(1,2,2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title("Training & Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend(["Train", "Validation"])

plt.show()
```

```python
model.save("plantvillage_cnn_model.h5")
print("Model saved as plantvillage_cnn_model.h5")
```

```python
from tensorflow.keras.models import load_model
import numpy as np
import cv2
import os
from google.colab import files
import matplotlib.pyplot as plt

# Load model
model = load_model("plantvillage_cnn_model.h5")

# Class names
DATASET_DIR = "/content/plantvillage/plantvillagedataset/color"
class_names = sorted(os.listdir(DATASET_DIR))

IMG_SIZE = (128, 128)

# Upload image
uploaded = files.upload()

for filename in uploaded.keys():

    # Read image
    img = cv2.imread(filename)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Preprocess
    img_resize = cv2.resize(img_rgb, IMG_SIZE)
    img_array = img_resize / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    # Predict probabilities
    pred = model.predict(img_array)[0]   # shape  (num_classes,)

    # Get top predictions
    top_3_idx = np.argsort(pred)[-3:][::-1]  # highest  lowest

    # Show image
    plt.imshow(img_rgb)
    plt.axis("off")
    plt.title("Predicted: " + class_names[np.argmax(pred)])
    plt.show()

    # Print full prediction result
    print(" Prediction Scores:")
    for i, prob in enumerate(pred):
        print(f"{class_names[i]}  {prob:.4f}")

    # Print Top-3 predictions
    print("\n Top 3 Predictions:")
    for idx in top_3_idx:
        print(f"{class_names[idx]}: {pred[idx]*100:.2f}%")

    print("\nFinal Prediction ", class_names[np.argmax(pred)])
```
